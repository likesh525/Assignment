# Abstract Class

Abstract class are blue prints used to create other classes. The class which uses abstract method is known as Abstract class. Abstract method can't implemented. 


```python
# Example
from abc import ABC, abstractmethod
class Vehicle():
    @abstractmethod
    def noofwheels(self):
        pass
        
class Truck(Vehicle):
    def noofwheel(self):
        print('4+ wheels')
        
class Bike(Vehicle):
    def noofwheels(self):
        print('2 wheels')
```

# Multiple inheritance

A class can be derived from more than one base class,  
In multilevel inheritance, features of the base class and the derived class are inherited into the new derived class.

```python
class A:
    print('in A')    
class B:
    print('in B')
    
class C(A,B):
    pass
    
obj1=C()
obj1
```
